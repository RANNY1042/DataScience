```python
#SVM을 이용한 타이타닉 생존자 분류(예측) 분석 수행

import numpy as np
import pandas as pd
import sklearn

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("C:/workspace_Bigdata/DataScience/PART 02 파이썬 분석/csv/titanic.csv")
df
```

```python
d_mean=df["Age"].mean()
df["Age"].fillna(d_mean,inplace=True)

d_mode=df["Embarked"].mode()[0]
df["Embarked"].fillna(d_mode,inplace=True)

df["FamilySize"]=df["SibSp"]+df["Parch"]
```

```python
onehot_sex=pd.get_dummies(df["Sex"])
df=pd.concat([df,onehot_sex],axis=1)
```

```python
onehot_embarked=pd.get_dummies(df["Embarked"])
df=pd.concat([df,onehot_embarked],axis=1)
```

```python
df
```

```python
X=df[["Pclass","Age","Fare","FamilySize","female","male","C","Q","S"]]
y=df["Survived"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel="linear",gamma=0.1)
sv.fit(X_train,y_train)
```

```python
pred=sv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc =accuracy_score(y_test, pred)
print(acc)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pred)
print(mat)
```

```python
from sklearn.metrics import classification_report
rpt = classification_report(y_test,pred)
print(rpt)
```
